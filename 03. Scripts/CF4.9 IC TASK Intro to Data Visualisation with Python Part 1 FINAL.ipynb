### Combining the two dfs

#### Checking if the user_id key is same on both dfs

```python
df_cust.info()
```

#### Merging two dfs with outer 

```python
# before i continue i need to drop column _merge from df_ords_prods_merge48 in order to merge the new dfs / saving the new format (no _merge column)

df_ords_prods_merge_drop=df_ords_prods_merge48.drop(columns = ['_merge'])
```

```python
# changing data type to save memory

df_ords_prods_merge_drop['user_id']=df_ords_prods_merge_drop['user_id'].astype(str)
```

```python
# performing downcast on df_ords_prods_merge_drop to reduce memory usage

for column in df_ords_prods_merge_drop:
    if df_ords_prods_merge_drop[column].dtype == "float64":
        df_ords_prods_merge_drop[column]=pd.to_numeric(df_ords_prods_merge_drop[column], downcast="float")
    if df_ords_prods_merge_drop[column].dtype == "int64":
        df_ords_prods_merge_drop[column]=pd.to_numeric(df_ords_prods_merge_drop[column], downcast="integer")
```

```python
# checking outcome

df_ords_prods_merge_drop.info()
```

```python
# merging two dfs

df_ords_prods_custs_final= pd.merge(df_ords_prods_merge_drop, df_cust, indicator = 'exists', on = 'user_id', how = 'outer')
```
