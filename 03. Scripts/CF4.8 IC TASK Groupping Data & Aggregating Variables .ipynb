# 6. The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:
If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

## I need to perform the following steps in order to create a new column with the avg_price for each customer
 1.Split the data into groups based on the “user_id” column
 2.Apply the transform() function on the “prices” column to generate the maximum orders for each user
 3.Create a new column , “avg_price” into which I’ll place the results of the aggregation

```python
df_ords_prods_merge['avg_price']=df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.max)
```

```python
df_ords_prods_merge.head(20)
```

## Deriving spending_flag column with loc function and checking outcome

```python
df_ords_prods_merge.loc[df_ords_prods_merge['avg_price'] >= 10, 'spending_flag']='High spender'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['avg_price'] < 10, 'spending_flag']='Low spender'
```

```python
df_ords_prods_merge['spending_flag'].value_counts()
```

```python
df_ords_prods_merge.head(20)
```

# 7. In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
df_ords_prods_merge['frequency_order_median']=df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.max)
```

```python
df_ords_prods_merge['frequency_order_median']
```

```python
df_ords_prods_merge.head(20)
```

## Deriving frequency_order_median_flag column with loc function and checking outcome


```python
df_ords_prods_merge.loc[df_ords_prods_merge['frequency_order_median'] > 20, 'frequency_order_median_flag']='Non-frequent customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['frequency_order_median'] > 10)&(df_ords_prods_merge['frequency_order_median'] <= 20), 'frequency_order_median_flag']='Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['frequency_order_median'] <= 10, 'frequency_order_median_flag']='Frequent customer'
```

```python
df_ords_prods_merge['frequency_order_median_flag'].value_counts()
```
