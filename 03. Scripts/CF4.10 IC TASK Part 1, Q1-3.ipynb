## Creating the For-Loop function for creating the 'region' column and checking outcome

```python
result_region = []
for value in df_ords_prods_custs_final['state']:
  if value in ['Maine','New Hampshire','Vermont','Massachusetts','Rhode Island','Connecticut','New York','Pennsylvania','New Jersey']:
    result_region.append("North East")
  elif value in ['Wisconsin','Michigan','Illinois','Indiana','Ohio','North Dakota','South Dakota','Nebraska','Kansas','Minnesota','Iowa','Missouri']:
    result_region.append("Midwest")
  elif value in ['Idaho','Montana','Wyoming','Nevada','Utah','Colorado','Arizona','New Mexico','Alaska','Washington','Oregon','California','Hawaii']:
    result_region.append("West")
  else:
    result_region.append("South")
```

### Adding result_region to a column in df and checking outcome

```python
df_ords_prods_custs_final['region']=result_region
```
